## __Create a linear regression model__

Here we use the [standard error](https://moderndive.com/10-inference-for-regression.html#regression-se) to calculate a confidence interval. Later we will use the bootstrap resampling technique
that does not rely on distrubutional assumptions.

```r
mod = linear_reg() |> set_engine('lm')

pen_model = mod |> fit(body_mass_g ~ flipper_length_mm, data = pen)

tidy(pen_model) |> mutate(lower_ci = estimate - 1.96 * std.error, upper_ci = estimate + 1.96 * std.error)
```

## __Question 2__

**Answer:**
>We are **95% confident** that the true effect of `flipper_length_mm` on `body_mass_g` lies within the range [46.71, 52.66]. <br>
>In other words, for each additional millimeter in **flipper length**, the **body mass** is expected to **increase by between 46.71 and 52.66** grams, on average.

## __Question 3__

Recall that the **test statistic** and **$p$-value** correspond to the
**hypothesis test**:

$$\begin{aligned} H_0:&\beta_{1} = 0 \\\ \mbox{vs }H_A:& \beta_{1} \neq 0 \end{aligned}$$

Write up the results & conclusions for this hypothesis test.

**Answer:**
>Based on the regression output, the **test statistic** is **32.72**, which is very large in magnitude, and the **p-value** is **4.37e-107**, an extremely small value approaching zero. Given that **p < 0.001**, the result is highly statistically significant.
>Since the p-value is far below conventional significance levels (e.g., 0.05 or 0.01), we can **reject the null hypothesis** ($H_0: \beta_1 = 0$), which states that flipper length has no effect on body mass.
>
>This provides **strong statistical evidence** that flipper length is significantly associated with body mass. Therefore, we can conclude that as **flipper length increases, body mass tends to increase** as well.

1.  Linearity of relationship between variables
2.  Independence of the residuals
3.  Normality of the residuals
4.  Equality of variance of the residuals

## __Question 5__

Now let’s circle back and take a second look at the confidence
intervals. Using this bootstrap distribution, we’ll construct the 95%

### __Step 1:__ Calculate the bootstrap statistic and visualize the bootstrap distribution.

```r
bootstrap_distn_slope = pen |> 
  specify(formula = body_mass_g ~ flipper_length_mm) |>
  generate(reps = 1000, type = "bootstrap") |>
  calculate(stat = "slope")

visualize(bootstrap_distn_slope)
```

### __Step 2:__ Calculate CI from the a bootstap resample using the percentile method.

Find a 95% CI using percentile method:

```r
bootstrap_distn_slope |>
    summarize(lower_ci = quantile(stat, 0.025), 
              upper_ci = quantile(stat, 0.975))
```

## __Question 10__

Run the code below to find the confidence intervals for our model's coefficients.

```r
boots = bootstraps(pen, times = 1000, apparent = TRUE)

mod = linear_reg() |> set_engine('lm')

boot_models  = boots |> 
  mutate(model = map(splits, ~mod |> fit(body_mass_g ~ bill_length_mm + species, data = as_tibble(.x))),
         coef_info = map(model, tidy))

boot_coefs = boot_models |> 
  unnest(coef_info)

percentile_intervals = int_pctl(boot_models, coef_info)

percentile_intervals
```

```r
boot_coefs |> 
    filter(term == 'bill_length_mm') |>
ggplot(aes(x = estimate)) +
    geom_histogram() +
    geom_vline(xintercept = percentile_intervals |> filter(term == 'bill_length_mm') |> pull(.lower), color = 'coral') +
    geom_vline(xintercept = percentile_intervals |> filter(term == 'bill_length_mm') |> pull(.upper), color = 'coral')
```

What can we conclude about the relationship between bill length and body
mass?

**Answer:**
> We have the 95% confidence intervals for the coefficients of the model based on bootstrapping. For `bill_length_mm`, the 95% confidence interval is between **78.90 and 103.71**. This means that, based on the bootstrapped samples, we are 95% confident that the true effect (coefficient) of bill length on body mass lies within this range. We can conclude that the relationship between bill length and body mass is **statistically significant and positive** because the confidence interval does not contain zero.
